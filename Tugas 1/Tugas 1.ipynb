## Library<a class="anchor" id="library"></a>
---

```python
# Importing library
import numpy as np
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from PIL import Image, ImageEnhance
import cv2
import math

import warnings 
warnings.filterwarnings('ignore')
```

## Nomor 1 <a class="anchor" id="nomor1"></a>
---

1. Dengan menggunakan citra **mountain.bmp**<br>
a. Transformasi menjadi citra biner menggunakan thresholding dengan nilai $T = 100$.

```python
# Read images
img_mountain = mpimg.imread('mountain.bmp')
img_mountain_biner = img_mountain.copy()

# Transform to binary image
img_mountain_biner[img_mountain_biner < 100] = 0
img_mountain_biner[img_mountain_biner >= 100] = 255

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_mountain, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_mountain_biner, cmap='gray')
plt.axis('off')
plt.title('Transformasi Citra Biner (T = 100)')
plt.show()
```

---
b. Transformasi menjadi citra negatif.

```python
# Read image
img_mountain = mpimg.imread('mountain.bmp')

# Transform to negative image
img_mountain_negative = 255 - img_mountain

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_mountain, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_mountain_negative, cmap='gray')
plt.axis('off')
plt.title('Transformasi Citra Biner (T = 100)')
plt.show()
```

---
c. Tingkatkan kecerahan citra dengan nilai $b = 100$.

```python
# Read image
img_mountain = Image.open('mountain.bmp')
enhancer = ImageEnhance.Brightness(img_mountain)
img_mountain_bright = enhancer.enhance(100)

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_mountain, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_mountain_bright, cmap='gray')
plt.axis('off')
plt.title('Citra yang dicerahkan (b = 100)')
plt.show()
```

## Nomor 2<a class="anchor" id="nomor2"></a>
---

2. Dengan menggunakan citra **girl-warna.bmp** konversi citra menggunakan nilai rata-rata RGB.

```python
# Read image
img_girlWarna = mpimg.imread('girl-warna.bmp')
img_girlWarna_convert = img_girlWarna.copy()

# Convert image using RGB average
img_girlWarna_convert = np.zeros((img_girlWarna.shape[0], img_girlWarna.shape[1]))
for i in range(img_girlWarna.shape[0]):
    for j in range(img_girlWarna.shape[1]):
        img_girlWarna_convert[i,j] = math.floor((img_girlWarna[i,j,0]+img_girlWarna[i,j,1]+img_girlWarna[i,j,2])/3)

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_girlWarna, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_girlWarna_convert, cmap='gray')
plt.axis('off')
plt.title('Convert using RGB average')
plt.show()
```

## Nomor 3<a class="anchor" id="nomor3"></a>

3. Kerjakan soal nomor 2 dengan menggunakan rumus **$Y = 0.299R + 0.587G + 0.144B$**

```python
# Read image
img_girlWarna = mpimg.imread('girl-warna.bmp')
img_girlWarna_convert2 = img_girlWarna.copy()

# Convert image using formula Y = 0.299R + 0.587G + 0.144B
img_girlWarna_convert2 = np.zeros((img_girlWarna.shape[0], img_girlWarna.shape[1]))
for i in range(img_girlWarna.shape[0]):
    for j in range(img_girlWarna.shape[1]):
        img_girlWarna_convert2[i,j] = math.floor((0.999*img_girlWarna[i,j,0])+(0.587*img_girlWarna[i,j,1])+(0.144*img_girlWarna[i,j,2]))

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_girlWarna, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_girlWarna_convert2, cmap='gray')
plt.axis('off')
plt.title('Convert using Y = 0.299R + 0.587G + 0.144B')
plt.show()
```

## Nomor 4<a class="anchor" id="nomor4"></a>

4. Gabungkan citra **bird.bmp** (citra A) dan **san.bmp** (citra B) menggunakan rumus **$C = \alpha A + (1 - \alpha)B$** dengan nilai **$\alpha = 0.8$**

```python
# Read images
imgA = mpimg.imread('bird.bmp')
imgB = mpimg.imread('san.bmp')

# Addition two images using formula C = A + (1 - )B with a value of  = 0.8
imgC = (0.8*imgA) + ((1-0.8)*imgB)

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(imgA, cmap='gray')
plt.axis('off')
plt.title('Citra A')
plt.subplot(132)
plt.imshow(imgB, cmap='gray')
plt.axis('off')
plt.title('Citra B')
plt.subplot(133)
plt.imshow(imgC, cmap='gray')
plt.axis('off')
plt.title('Citra A + Citra B ('+ r'$\alpha = 0.8$'+')')
plt.show()
```

## Nomor 5<a class="anchor" id="nomor5"></a>
---

5. Tunjukkan apakah terdapat perbedaan antara citra **lena_gray.bmp** dan **lena_gray.gif**.

```python
# Read images
img_lena_bmp = mpimg.imread('lena_gray.bmp')
img_lena_gif = mpimg.imread('lena_gray.gif')

# Convert lena_gray.gif to grayscale
img_lena_gif2 = cv2.cvtColor(img_lena_gif, cv2.COLOR_BGR2GRAY)

# Subtraction two images
img_subtract = cv2.subtract(img_lena_bmp, img_lena_gif2)

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(img_lena_bmp, cmap='gray')
plt.axis('off')
plt.title('lena_gray.bmp')
plt.subplot(132)
plt.imshow(img_lena_gif, cmap='gray')
plt.axis('off')
plt.title('lena_gray.gif')
plt.subplot(133)
plt.imshow(img_subtract, cmap='gray')
plt.axis('off')
plt.title('Subtraction')
plt.show()
```

Perbedaanya hanya pada tingkat kecerahan dan ketajaman citra. Citra **lena_gray.gif** di convert ke grayscale agar memiliki dimensi _shape_ (ukuran) yang sama dengan citra **lena_gray.bmp** sehingga dapat dilakukan _subtraction_ (pengurangan) terhadap dua citra untuk melihat perbedaan diantara keduanya.

6. Tingkatkan kecerahan citra **zelda.bmp** dengan menggunakan perkalian dengan konstanta $k = 2$.

```python
# Read image
img_zelda = mpimg.imread('zelda.bmp')

# Increase image brightness using multiplication by constant k = 2
img_zelda_bright = 2 * img_zelda

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_zelda, cmap='gray')
plt.axis('off')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_zelda_bright, cmap='gray')
plt.axis('off')
plt.title('Citra yang dicerahkan (k=2)')
plt.show()
```

## Nomor 7<a class="anchor" id="nomor7"></a>

7. Dengan menggunakan **bird.bmp** dan **san.bmp** tampilkan hasil operasi logika boolean **or** terhadap dua citra gray scale tersebut.

```python
# Read images
img_bird = mpimg.imread('bird.bmp')
img_san = mpimg.imread('san.bmp')

# Copying images (defining variable for binary image)
img_bird_biner = img_bird.copy()
img_san_biner = img_san.copy()

# Transform bird.bmp to binary image
img_bird_biner[img_bird_biner < 128] = 1
img_bird_biner[img_bird_biner >= 128] = 0

# Transform san.bmp to binary image
img_san_biner[img_san_biner < 128] = 1
img_san_biner[img_san_biner >= 128] = 0

# Boolean operation OR
img_or = img_bird.copy()
for i in range(img_bird_biner.shape[0]):
    for j in range(img_bird_biner.shape[1]):
        img_or[i,j] = img_bird_biner[i,j] or img_san_biner[i,j]

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(img_bird_biner, cmap='gray')
plt.axis('off')
plt.title('Binary image bird.bmp')
plt.subplot(132)
plt.imshow(img_san_biner, cmap='gray')
plt.axis('off')
plt.title('Binary image san.gif')
plt.subplot(133)
plt.imshow(img_or, cmap='gray')
plt.axis('off')
plt.title('logika boolean or')
plt.show()
```

9. Cerminkan citra **lena_color.gif** terhadap titik asal.

10. Perbesar citra **girl-warna.bmp** menjadi $512 \times 512$

```python
# Read image
img_girlWarna = Image.open('girl-warna.bmp')

# Resize girl-warna.bmp to 512 x 512
img_girlWarna_resize = img_girlWarna.resize((512, 512), Image.BILINEAR)

# Displaying images
plt.figure(figsize=(15,5))
plt.subplot(121)
plt.imshow(img_girlWarna, cmap='gray')
plt.title('Citra Asli')
plt.subplot(122)
plt.imshow(img_girlWarna_resize, cmap='gray')
plt.title('Citra diperbesar '+ r'$512 \times 512$')
plt.show()
print("Sebelum",img_girlWarna.size)
print("Sesudah",img_girlWarna_resize.size)
```

## Library
---

```python
import numpy as np
import numpy.fft as npfp
import matplotlib.pylab as pylab
from scipy import fftpack
from matplotlib.colors import LogNorm
```

## Soal
---

Hilangkan noise periodik pada gambar __lena_priodic.bmp__

## Original Image
---

```python
img = pylab.imread('lena_priodic.bmp')
pylab.figure(figsize=(10, 10))
pylab.imshow(img, cmap='gray')
pylab.axis('off')
pylab.title('Periodic Noise Image (Original Image)')
```

## Fourier Transform
---

```python
img_fft = fftpack.fft2(img)

img_fftshift = fftpack.fftshift(np.abs(img_fft))

pylab.figure(figsize=(10,10))
pylab.imshow(img_fftshift, norm=LogNorm(vmin=5), cmap='gray')
pylab.colorbar()
pylab.title('Spectrum Fourier transform', size=15)
```

```python
keep_fraction = 0.1
img_fft2 = img_fft.copy()

# Set r and c sebagai jumlah baris dan kolom array
r, c = img_fft2.shape

# Set semua baris ke nol dengan indeks antara r * keep_fraction dan r * (1 - keep_fraction)
img_fft2[int(r * keep_fraction):int(r * (1 - keep_fraction))] = 0

# Set semua kolom ke nol dengan indeks antara r * keep_fraction dan r * (1 - keep_fraction)
img_fft2[:, int(c * keep_fraction):int(c * (1 - keep_fraction))] = 0

img_fftshift2 = fftpack.fftshift(np.abs(img_fft2))
pylab.figure(figsize=(10,10))
pylab.imshow(img_fftshift2, norm=LogNorm(vmin=5), cmap='gray')
pylab.colorbar()
pylab.title('Filtered Spectrum', size=15)
```

## Reconstructed Image
---

```python
img_recons = npfp.ifft2(img_fft2).real
pylab.figure(figsize=(10,10))
pylab.imshow(img_recons, cmap='gray')
pylab.axis('off')
pylab.title('Reconstructed Image', size=15)
```
